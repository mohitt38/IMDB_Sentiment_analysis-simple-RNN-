```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
## sentences
sent=[ 'the glass of milk', 
    'the glass of juice',
    'the cup of tea',
    'I am good boy',
    'I a good developer',
    'understand the meaning of words',
    'your videos are good',]
```

```python
## define the vocabllury size
voc_size = 10000
```

```python
sent
```

```python
## One hot Representation
one_hot_repr=[one_hot(words,voc_size)for words in sent]
one_hot_repr

```

```python
## word embedding representation

from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
sent_len = 8
embedding_docs=pad_sequences(one_hot_repr,padding='pre',maxlen=sent_len)
print(embedding_docs)
```

```python
## feature represntation
dim=10
```

```python
model = Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_len))
model.compile('adam', 'mse')
```

```python
model.summary()
```

```python
model.predict(embedding_docs)
```
